```python
from openai import OpenAI
```

```python
openai = OpenAI()
```

```python
messages = [
    {"role": "user", "content": " what is 2 + 2?"}
]
```

```python
responce = openai.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=messages
)
print(responce.choices[0].message.content)
```
